```python
# Strip spaces and replace inner spaces with underscores for all column names
df.columns = df.columns.str.strip().str.replace(r'\s+', '_',regex=True)
```

```python
df.columns = df.columns.str.replace(r'[()\s]', '', regex=True).str.replace(r'__+', '_')
```

```python
# Identify columns
area_cols = [col for col in df.columns if 'AREA' in col.upper()]
production_cols = [col for col in df.columns if 'PRODUCTION' in col.upper()]
yield_cols = [col for col in df.columns if 'YIELD' in col.upper()]
```

```python
# Convert Area and Production from 1000 ha/tons to ha/tons
for col in area_cols:
    df[col] *= 1000
    print(f"📐 Standardized '{col}' to hectares (ha)")
```

```python
for col in production_cols:
    df[col] *= 1000
    print(f"⚖ Standardized '{col}' to tons")
```

```python
# Validate Yield: Production / Area = Yield (Kg/ha)
for prod_col, area_col, yield_col in zip(production_cols, area_cols, yield_cols):
    df[f"calculated_{yield_col}"] = (df[prod_col] / df[area_col]).replace([float('inf'), -float('inf')], 0).fillna(0)
    diff = (df[yield_col] - df[f"calculated_{yield_col}"]).abs().max()
    print(f"✅ Yield Check '{yield_col}' | Max difference: {diff:.2f} Kg/ha")
```

```python
# Preview
print("\n🔍 Standardized Preview:")
print(df[[area_cols[0], production_cols[0], yield_cols[0], f"calculated_{yield_cols[0]}"]].head())
```

1.Year-wise Trend of Rice Production Across States (Top 3)

```python
cursor.execute("""SELECT year, state_Name, SUM(rice_production_1000_tons) AS Total_rice_Production
FROM agricultural_data
WHERE state_Name IN (
    SELECT state_Name
    FROM agricultural_data
    GROUP BY state_Name
    ORDER BY SUM(rice_production_1000_tons) DESC
    LIMIT 3
)
GROUP BY year, state_Name
ORDER BY year, Total_rice_Production DESC;

 
 """)
out=cursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in cursor.description],  tablefmt='psql'))
```

2.Top 5 Districts by Wheat Yield Increase Over the Last 5 Years

```python
cursor.execute("""SELECT dist_name, 
       (MAX(WHEAT_YIELD_Kg_per_ha) - MIN(WHEAT_YIELD_Kg_per_ha)) AS Yield_Increase
FROM agricultural_data
WHERE Year >= (SELECT MAX(Year) FROM agricultural_data) - 5
GROUP BY dist_name
ORDER BY Yield_Increase DESC
LIMIT 5;
 """)
out=cursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in cursor.description],  tablefmt='psql'))
```

3.States with the Highest Growth in Oilseed Production (5-Year Growth Rate)

```python
cursor.execute("""SELECT State_Name,
       COALESCE(((MAX(oilseeds_production_1000_tons) - MIN(oilseeds_production_1000_tons)) / NULLIF(MIN(oilseeds_production_1000_tons), 0)) * 100, 0) AS Growth_Rate_Percentage
FROM agricultural_data
WHERE Year BETWEEN (SELECT MAX(Year) - 4 FROM agricultural_data) AND (SELECT MAX(Year) FROM agricultural_data)
GROUP BY State_Name
HAVING COUNT(DISTINCT Year) = 5
ORDER BY Growth_Rate_Percentage DESC
LIMIT 5;


 """)
out=cursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in cursor.description],  tablefmt='psql'))
```

4.District-wise Correlation Between Area and Production for Major Crops (Rice, Wheat, and Maize)

```python
cursor.execute("""WITH Top5CottonStates AS (
    SELECT state_name
    FROM agricultural_data
    GROUP BY state_name
    ORDER BY SUM(cotton_production_1000_tons) DESC
    LIMIT 5
),
YearlyGrowth AS (
    SELECT state_name,
           year,
           ((cotton_production_1000_tons - LAG(cotton_production_1000_tons) OVER (PARTITION BY state_name ORDER BY year)) /
           LAG(cotton_production_1000_tons) OVER (PARTITION BY state_name ORDER BY year)) * 100 AS growth_rate_percentage
    FROM agricultural_data
    WHERE state_name IN (SELECT state_name FROM Top5CottonStates)
)
SELECT year, 
       ROUND(AVG(growth_rate_percentage), 2) AS average_growth_rate
FROM YearlyGrowth
WHERE growth_rate_percentage IS NOT NULL
GROUP BY year
ORDER BY year;

 """)
out=cursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in cursor.description],  tablefmt='psql'))
```
